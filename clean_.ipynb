```python
#import semua library yang dibutuhkan
import numpy as np
import pandas as pd
import json
import ast
pd.options.display.max_columns=30
```

```python
# Load data
df = pd.read_csv('movies_metadata.csv', low_memory=False)
df.head()
```

```python
#melihat informasi dataset
df.info()
```

```python
# Cleaning data number dan text menjadi number
df[['revenue','runtime','vote_average','vote_count']]
```

```python
#menjadikan budget serta revenue menjadi numeric dan mejadikan satuannya jutaan
df.budget = pd.to_numeric(df.budget, errors='coerce')
```

```python
df.budget = df.budget.replace(0, np.nan)
df.budget = df.budget.div(1000000)
df.budget
```

```python
df.revenue=df.revenue.astype('float').replace(0,np.nan)
df.revenue = df.revenue.div(1000000)
df.revenue
```

```python
df = df.rename(columns={'budget':'budget_musd','revenue':'revenue_musd'})
```

```python
df.runtime.value_counts()
```

```python
#menjadikan value 0 menjadi NaN
df.runtime = df.runtime.replace(0, np.nan)
```

```python
#menjadikan ID jadi numeric
df.id = pd.to_numeric(df.id, errors='coerce')
```

```python
df.popularity = pd.to_numeric(df.popularity, errors='coerce')
```

```python
df.loc[df.vote_count== 0, 'vote_average']
```

```python
df.loc[df.vote_count== 0, 'vote_average']= np.nan
```
