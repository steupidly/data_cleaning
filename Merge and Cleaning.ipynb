Merging Data (left join)

```python
#check apakah ada credit.id di df.id
df[~df.id.isin(credit.id)]
```

```python
#check apakah ada df.id di credit.id
credit[~credit.id.isin(df.id)]
```

```python
#merge(dataframe kanan, join apa, relasi kiri, relai kanan)
df = df.merge(credit, how="left", left_on="id", right_on="id")
```

```python
df.info()
```

Cleaning new column cast

```python
df.cast[0]
```

```python
df.cast = df.cast.apply(lambda x : ast.literal_eval(x)if isinstance (x,str)else np.nan)
```

```python
pd.DataFrame(df.cast[0])
```

```python
#mebuat colom baru berapa banyak castnya dalam 1 film
df['cast_size']= df.cast.apply(lambda x: len(x))
```

```python
df.cast_size.value_counts(dropna=False)
```

```python
#hanya menggabil nama aktornya saja
df.cast= df.cast.apply(lambda x : "|".join(i['name']for i in x)if isinstance (x,list)else np.nan)
```

```python
df.cast.value_counts().head(50)
```

```python
#drop missing value
df.cast.replace("",np.nan,inplace=True)
```

```python
df.head()
```

Cleaning new column crew

```python
df.crew[0]
```

```python
df.crew.apply(lambda x : ast.literal_eval(x)if isinstance(x,str)else np.nan)
```

```python
df.crew = df.crew.apply(lambda x : ast.literal_eval(x)if isinstance(x,str)else np.nan)
```

```python
pd.DataFrame(df.crew[0])
```

```python
df['crew_size']= df.crew.apply(lambda x : len(x))
```

```python
df.crew_size.value_counts(dropna=False)
```

```python
#mencari nama director di suatu film
def get_director(x):
    for i in x:
        if i['job'] == 'Director':
            return i['name']
    return np.nan

```

```python
#buat colomn director berisi nama-namanya
df['director']=df.crew.apply(get_director)
```

```python
df.director.value_counts(dropna=False)
```

```python
df.drop(columns=['crew'], inplace=True)
```
